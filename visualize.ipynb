```python
import os
from pathlib import Path

DATA_DIR = os.path.join(Path(os.path.curdir).parent, 'data')

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def visualize(df_corr):
    """
    히트맵 시각화하는 함수
    :param df_corr:
    :return:
    """
    colormap = plt.cm.PuBu
    plt.figure(figsize=(12, 10))
    plt.title("Correlation")
    sns.heatmap(
        df_corr.astype(float).corr(),
        linewidths=0,
        square=False,
        cmap=colormap,
        linecolor="white",
    )
    plt.show()
```

```python
corr_file_name = 'ml_corr_map.csv'
csv = pd.read_csv(os.path.join(DATA_DIR, corr_file_name))
```

```python
visualize(csv)
```
